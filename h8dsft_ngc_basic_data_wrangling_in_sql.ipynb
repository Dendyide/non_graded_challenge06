retrieve all trips where the bike_id is not null and greater than 1000 minutes

```python
query_bike_trips = """
SELECT bike_id, SUM(duration_minutes) AS duration_bike
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
GROUP BY bike_id
HAVING SUM(duration_minutes) > 1000
ORDER BY duration_bike DESC;
"""
```

```python
query_job_bike_trips = client.query(query_bike_trips).to_dataframe()
query_job_bike_trips
```

filter trip_id is "Walk Up" and end_station_name is "Stolen"

```python
query_bike_filter = """
SELECT trip_id
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
WHERE subscriber_type LIKE '%Walk Up%' and end_station_name LIKE '%Stolen%'
"""
```

```python
query_job_bike_filter = client.query(query_bike_filter).to_dataframe()
query_job_bike_filter
```
